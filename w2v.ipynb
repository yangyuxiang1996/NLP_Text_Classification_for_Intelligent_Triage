```python
import joblib
from gensim import models


tfidf = joblib.load('ml/model/tfidf')
w2v = models.KeyedVectors.load_word2vec_format(
            './DL/data/w2v_1.bin', binary=True)
lda = models.ldamodel.LdaModel.load('ml/model/lda')
```

```python
import numpy as np
label2id = {'男科': '1', '皮肤科': '2', '其它': '3', '妇科': '4', '心血管内科': '5', '耳鼻喉科': '6', '消化内科': '7', '小儿内科': '8', '呼吸内科': '9', '内分泌科': '10', '神经内科': '11'}
w2v_label_embedding = np.array([np.mean([w2v.get_vector(word) for word in key if word in w2v.key_to_index],  axis=0) for key in label2id])
w2v_label_embedding[0]
```

```python
w2v_label_embedding = np.zeros(shape=(len(label2id), w2v.vector_size))
for i, key in enumerate(label2id.keys()):
    vec = np.zeros(w2v.vector_size)
    for word in key:
        if word in w2v.key_to_index:
            vec += w2v[word]
    vec /= len(key)
    w2v_label_embedding[i] = vec

w2v_label_embedding.shape
```
